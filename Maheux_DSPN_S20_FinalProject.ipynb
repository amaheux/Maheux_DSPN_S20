# Title: The Timing and Context of "Coming Out" for Depressive Symptoms among U.S. Transgender Adolescents


# Background


Adolescents identifying as transgender (i.e., those that do not identify with their gender assigned at birth) are gaining visibility and attention from researchers. In 2019, the CDC reported that nearly 2 percent of high school students identified as transgender (CDC, 2019). 

Transgender adolescents experience depressive disorders at six times the rate of cisgender adolescents, with about half of all transgender adolescents meeting diagnostic criteria for major depressive disorder (Becerra-Culqui et al., 2018; Reisner et al., 2016). In addition to the reduced psychological well-being among this popultion, the disparity in depressive symptoms underlies the higher rates of suicidality (Thoma et al., 2019), anxiety (Durwood et al., 2017), non-suicidal self-injury (Duffy et al., 2019), and worse body satisfaction (Jones et al., 2015) among transgender teens. Minority stress theory (Meyer, 2003) contends that the expected and experienced stigma, prejudice, and discrimination members of minority groups face creates a stressful social environment that contributes to mental health problems. 

Among these minority stressors, transgender adolescents face numerous social stressors related to "coming out," or disclosing their identity to others (Klein et al., 2015). Adolescents must navigate not only when to come out and to whom, but the broader sociopolitical context of their social environments that may affect the reactions they receive to coming out. It remains unclear how experiences of coming out, including the timing, proximal factors, and distal contextual factors related to coming out, affect transgender adolescents' depressive symptoms. The present study investigates the association between timing of coming out (the duration between wondering about one's true gender identity and first telling someone and the duration since coming out) and depressive symptoms. The study will also examine the role of proximal and distal social and sociopolitical factors in predicting adolescents' depressive symptoms.

# Variables

All variables were collected via self-report surveys. Adolescents were recruited through social media advertisements for a broader study on transgender adolescents' health and well-being. Self-report surveys were anonymous. All participants were between the ages of 14-18.

### Outcome variable:

Depressive symptoms (depressSUM; quantitative): adolescents completed the CES-D, a 20-item scale that measures depressive symptoms in the past week on a scale from 0-3. Items were summed with higher scores indicating greater depressive symptoms and a maximum score of 60. The scale shows good psychometric properties with adolescent samples (Mojarrad & Lennings, 2002).

### Demographic variables:

Gender identity (transmasc, transfem, nonbinfem, nonbinmale, questioning; qualitative): adolescents reported their gender identity by indicating one or more from a comprehensive list of gender identities. Adolescents were categorized into 5 groups: (1) transfemale/transfeminine, (2) transmale/transmasculine, (3) nonbinary assigned female at birth, (4) nonbinary assigned male at birth, and (5) questioning their gender identity. Identity groups have been dummy coded.

Age (Age; quantitative): adolescents reported their numeric age in years

Race (white; qualitative): adolescents indicate one or more racial groups from a list. For parsimony race has been dichotomized with 1 = white, 0 = racial minority. 

### Timing of coming out variables:

Duration between first wondering about true gender identity and first telling someone (TimeWonderingAlone; quantitative): participants indicated their age in years at which they first wondered about their true gender minority identity and their age in years at which they first told someone ("came out"). The duration between those points in time was calculated by taking the difference score (age came out - age first wondered).

Age since first telling someone of true gender identity (TimeOut; quantitative): participants indicated their age in years at which they first told someone about their gender identity (age they "came out"). The duration between when they first came out and the time at which they completed the survey was calculated by taking the difference score between their current age and the age they first came out (current age - age came out).


### Proximal social context of coming out variables:

Have come out to at least one person (ComeOut; qualitative): Participants indicated if they had ever told anyone of their true gender identity. Responses were dummy coded such that 1 = has come out, 0 = has not come out.

Who knows about true gender identity (FamKnows, FriendsKnow, SibsKnow, OtherRelKnows, TeacherKnows, ClassmatesKnow, NotComeOut; qualitative): participants were presented with a list of social connections (i.e., family, friends, siblings, other relatives, teacher, and classmates) and indicated who currently knows about their true gender identity. All connections were dummy coded so that 1 = that connection knows and 0 = that connection doesn't know. 

Family relationship quality (famwarmth; quantitative): participants indicated the extent to which their family is warm and supportive with 4-items measured on a scale from 1 to 5. Mean scores were calculated such that higher scores indicated greater familial warmth and relational quality.

Number of close friends (CloseFriends; quantitative): participants indicated their number of close friends on a scale from 1 to 5 with 1 = none, 2 = 1 or 2, 3 = 3-5, 4 = 6-9, 5 = 10 or more.

Number of LGBT friends (LGBTfriends; quantitative): participants indicated their number of LGBT friends on a scale from 1 to 5 with 1 = none, 2 = 1 or 2, 3 = 3-5, 4 = 6-9, 5 = 10 or more.

Affiliation with other LGBT adolescents (LGBTclub, LGBTclub_welcoming; qualitative): adolescents reported on whether or not their school has an LGBT club (e.g., Gay-Straight Alliance) and if that club is welcoming. Participants indicated "yes" or "no." Both items were dummy coded.

Exposure to transgender adolescents (TransTeensCommunity, TransTeensOnline, TransTeensSchool; quantitative): adolescents reported the number of transgender teens in their local community, that they interact with regularly online, and at their school. 



### Distal sociopolitical context of coming out variables:

Region (Northeastdum, Southdum, Westdum, Midwestdum; qualitative): participants indicated their zipcode. National zipcode data was used to identify states that participants lived in. Regions were categorized based on Census regional categorizations. All region variables were dummy coded.

State population (Population_2019estimate; quantitative): participant zipcode data was linked with data from the US census that estimated 2019 state population to identify the population of the state in which the participant resides.

Political leaning of state (reddummy, bluedummy, purpledummy; qualitative): Presidential elections from the past 4 cycles (2004, 2008, 2012, and 2016) were used to identify political leanings of the U.S. states. States that voted for Democrats all four years were coded as blue, states that voted for Republicans all four years were coded as red, all other states were coded as purple. Participants' zipcodes were used to link their state of residence with the state's  political leaning. All political leaning variables were dummy coded.









## Hypothesis

In regards to the timing of coming out, I predict that a longer duration between first wondering about one's true gender identity and first coming out will be associated with more depressive symptoms and that a longer duration since coming out will be associated with less depressive symptoms. Specifically, I predict that, using linear regression, the coefficient for the TimeWonderingAlone variable will be positive and the TimeOut variable will be negative. I predict that, using a permuted null distribution, the probability of these coefficients being observed given that null will be less than .05. In addition, I predict to find evidence for these models using Bayes Factors and to find evidence against the null. 

In regards to the social and political context of coming out, I predict that, using random forest to manage correlations between predictors, I will be able to predict the depressive symptoms of new observations better than with the prior timing variables only. In other worse, the random forest model with all the proximal and distal contextual factors will yield a MSE from the hold-out test set that will be lower compared to the MSE of the hold-out set for the linear regression models. I predict that proximal predictors (such as familial warmth, who knows of the participants' true gender identity, and number of close friends) will be more important in predicting depressive symptoms than distal predictors (such as region or political leaning).

## (1) The association between timing of coming out and depressive symptoms

```r
#linear regression for time spent wondering alone
model1 = lm(depressSUM~TimeWonderingAlone + Age, data=finaldataTiming)
summary(model1)
#checking assumptions of OLS regression
plot(model1)
#although there are a few outliers, the data fit the assumptions of linearity, independence, and normal distribution of Y
```

```r
#linear regression for time since coming out
model1 = lm(depressSUM~TimeOut + Age, data=finaldataTiming)
summary(model1)
#checking assumptions of OLS regression
plot(model1)
#although there are a few outliers, the data fit the assumptions of linearity, independence, and normal distribution
```

```r
#Graph for TimeWonderingAlone
library(ggplot2)

finaldataTiming$TimeWonderingAlone <- as.numeric(finaldataTiming$TimeWonderingAlone)
finaldataTiming$depressSUM <- as.numeric(finaldataTiming$depressSUM)

ggplot(finaldataTiming, aes(x=TimeWonderingAlone, y=depressSUM)) + geom_point() + geom_smooth(method=lm) + labs(title = "Association between Time Wondering Alone and Depressive Symptoms", y= "Depressive Symptoms", x = "Duration Between First Wondering about True Gender Identity and First Telling Someone")

#Graph for TimeOut
library(ggplot2)

finaldataTiming$TimeOut <- as.numeric(finaldataTiming$TimeOut)

ggplot(finaldataTiming, aes(x=TimeOut, y=depressSUM)) + geom_point() + geom_smooth(method=lm) + labs(title = "Association between Time Since Coming Out and Depressive Symptoms", y= "Depressive Symptoms", x = "Time Since First Came Out")

```

```r
# permutation test for TimeWonderingAlone null distribution

permuteddata <- finaldataTiming
R=1000
perm.coefs=matrix(NA,nrow=R, ncol=2)

for (i in 1:R){
  permuteddata$TimeWonderingAlone=finaldataTiming$TimeWonderingAlone[sample(864)]
  perm.coefs[i,]=coef(lm(depressSUM~TimeWonderingAlone, data=permuteddata))
}
realmodel = coef(lm(depressSUM~TimeWonderingAlone  + Age, data=finaldataTiming))
perm.p = sum(perm.coefs[,2]>realmodel[2])/R 
perm.p

#probability of observing an effect greater than the observed effect given the permuted null distribution is < .0001


# permutation test for TimeOut null distribution

permuteddata <- finaldataTiming
R=1000
perm.coefs=matrix(NA,nrow=R, ncol=2)

for (i in 1:R){
  permuteddata$TimeOut=finaldataTiming$TimeOut[sample(864)]
  perm.coefs[i,]=coef(lm(depressSUM~TimeOut, data=permuteddata))
}
realmodel = coef(lm(depressSUM~TimeOut  + Age, data=finaldataTiming))
perm.p = sum(perm.coefs[,2]>realmodel[2])/R 
perm.p

#probability of observing an effect greater than the observed effect given the permuted null distribution is .02
```

```r
#Cross-validation for TimeWonderingAlone: predicting depressive symptoms among new observations
set.seed(3)
train = sample(1: nrow(finaldataTiming), nrow(finaldataTiming)/2)
model1 = lm(depressSUM~TimeWonderingAlone + Age, data=finaldataTiming, subset=train)

yhat =predict(model1, newdata=finaldataTiming[-train,])

testvalues=finaldataTiming[-train, "depressSUM"]
mean((yhat-testvalues)^2)
sqrt(mean((yhat-testvalues)^2))

plot(yhat ,testvalues)
abline(0,1)
# cross-validation finds a MSE of 76.11 for this model on the hold-out test set. 
#Square root of the MSE is 8.72, indicating that the model is off by, on average, 
#8.72 points on the depressive symptoms scale
#when predicting the depressive symptoms of the observations in the hold-out test set.



#Cross-validation for TimeOut: predicting depressive symptoms among new observations
train = sample(1: nrow(finaldataTiming), nrow(finaldataTiming)/2)
model1 = lm(depressSUM~TimeOut + Age, data=finaldataTiming, subset=train)

yhat =predict(model1, newdata=finaldataTiming[-train,])

testvalues=finaldataTiming[-train, "depressSUM"]
mean((yhat-testvalues)^2)
sqrt(mean((yhat-testvalues)^2))

plot(yhat ,testvalues)
abline(0,1)
# cross-validation finds a MSE of 82.28 for this model on the hold-out test set. 
#Square root of the MSE is 9.07, indicating that the model is off by, on average, 
#9.07 points on the depressive symptoms scale
#when predicting the depressive symptoms of the observations in the hold-out test set.
```

```r
#install.packages('BayesFactor')
library(BayesFactor)

TimeWonderingmodel <- lmBF(depressSUM ~ TimeWonderingAlone + Age, data=finaldataTiming) 
TimeOutmodel <- lmBF(depressSUM ~ TimeOut + Age, data=finaldataTiming) 
Null <- lmBF(depressSUM ~ Age, data=finaldataTiming)

allBFs <- c(TimeWonderingmodel, TimeOutmodel, Null)
allBFs

1/TimeWonderingmodel
1/TimeOutmodel
TimeWonderingmodel/TimeOutmodel


#With a Bayes Factor of 256.73, there is very strong evidence for the 
#alternative hypothesis of the association between TimeWonderingAlone and depressive symptoms.
#The Bayes factor for the null is very small, presenting evidence against the null hypothesis

#In contrast, with a Bayes Factor of 0.11, there is very weak evidence for the 
#alternative hypothesis of the association between TimeOut and depressive symptoms. 
#For this model, the BF01 is 9.12, suggesting "positive" evidence for the null 

#The Bayes factor analysis also shows the comparison of the TimeWondering model to the TimeOut model.
#In this comparison, the TimeWondering model has a BF10 of 2342, providing very strong evidence for 
#the TimeWondering model relative to the TimeOut model.
```

Summary:
There is strong evidence for an association between the duration between wondering about one's true gender identity and first telling someone and an individuals' depressive symptoms. Linear regression shows that, controlling for age, this duration is significantly positively associated with more depressive symptoms. The graph plots the raw data points and shows the positive slope of the regression line. Using a permutation test to determine the null distribution, the probability of finding the observed coefficient for TimeWonderingAlone is less than .0001. The very large Bayes Factor for the alternate model indicates strong evidence of the alternate model and evidence against the null model. Cross-validation shows the error associated in predicting the depressive symptoms of observations in a hold-out test set is off, on average, by about 8-9 points.

The evidence for an association between the duration of time since coming out and depressive symptoms is unclear. The graph plots the raw data points and shows the slope of the regression line, which is slightly positive though the data is clustered and thus confidence of estimates as X increases reduces. Linear regression shows that, controlling for age, this duration is significantly positively associated with more depressive symptoms. When determining the null distribution with a permutation test, we see that the probability of finding the coefficient for TimeOut that was determined by these data is about .02. The very small Bayes Factor for the alternate model indicates evidence against the alternate model and the medium-sized Bayes Factor for the null suggests some positive evidence for the null model. Similar to the previous model, cross-validation shows the error associated in predicting the depressive symptoms of observations in a hold-out test set is off, on average, by about 8-9 points.

Taken together, these results suggest a strong link between the duration between wondering about one's gender identity and first coming out and depressive symptoms, but conflicting evidence for the associaiton between time since coming out and depressive symptoms. However, the cross-validated error is similar, suggesting that neither model is better at predicting depressive symptoms in a new sample. The following analyses leverage more variables to attempt to predict the depressive symptoms of a hold-out test set using a random forest.

## (2) The association between social and sociopolitical context and depressive symptoms

```r
#install.packages('randomForest')
library(randomForest)

set.seed (2)
train = sample(1: nrow(finaldataSociopol), nrow(finaldataSociopol)/2)

forest.model=randomForest(depressSUM~., data=finaldataSociopol, subset=train, mtry=3, ntree=5000,
                          importance=TRUE)
forest.model

testsetvalues=finaldataSociopol[-train, "depressSUM"]
yhat.rf = predict(forest.model, newdata=finaldataSociopol[-train ,])

mean((yhat.rf-testsetvalues)^2)
sqrt(mean((yhat.rf-testsetvalues)^2))

importance(forest.model)
varImpPlot(forest.model)
```

Summary:

On the hold-out test set, the random forest model has a MSE of 77.16, which is not better than the linear regression model with two predictor variables. However, the random forest model explains 18.74% of the overall variance in depressive symptoms, which is an increase from the previous linear regression models. Moreover, the current model tells us something about the social and sociopolitical context of coming out as a transgender adolescent.

The variable importance plot shows that famwarmth (or familial warmth and relational quality), followed by TimeWonderingAlone and number of close friends, are most important in reducing the MSE. When these variables are not included in the tree, the decrease in the accuracy of predictions is, on average, 45 points for famwarmth, 27 points for TimeWonderingAlone, and 24 points for number of close friends. These decreases in prediction accuracy are substantial compared to the overall MSE of the random forest model and suggest that these variables are highly relevant in predicting the depressive symptoms of transgender adolescents.


## Conclusion

These models show that some timing and social context variables are more important than others in predicting depressive symptoms among transgender adolescents. Telling another person about one's true gender identity soon after an adolescent wonders about their identity and having close others (e.g., a warm family and many close friends) with whom to discuss the process may be most important in determining adolescents' depressive symptoms. Although other distal factors (e.g., duration since coming out, political leaning of state of residence) may play a slight role, these variables do not directly measure the quality of an adolescents' interpersonal relationships or time spent without receiving support from those relationships. Importantly, these results suggest that all adolescents--regardless of state of residence or the broader context of their school or community--may be at risk of elevated depressive symptoms if they do not have close others with whom they can disclose their identities and reap support. The provision of social support by close friends and family likely underlies these findings and may be the most important factor in determining, and also potentially mitigating, the depressive symptoms of this vulnerable group.
