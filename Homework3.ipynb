1. How does knowing data type (e.g., qualitative vs. quantitative) provide insights into the type of distribution a random variable comes from?

Different types of data come from different distributions, which can give us insight into the probability of getting any particular value of a variable when a data point is chosen at random from its distribution. Quantitative variables come from continuous probability distributions (e.g., the normal curve/gaussian)  whereas qualitative variables come from frequency distributions (e.g., binomial distribution).

To say that a random variable is normal with a mean of 100 and a standard deviation of 15 means that we can assume the probability distribution of the variable is symmetrical. For this variable, 68.3% of the data points are between 85 and 115 and  95.4% of the data points are between 70 and 130. 

The probability of pulling a random data point at the mean will be highest, while the probability of pulling a random data point with a value above or below the mean will decrease exponentially as the value moves farther away from the mean

4. Instead of rolling one die, run a set of experiments reporting the outcomes of rolling multiple dice at the same time.
        	a) Exp 1: 10,000 throws, 1 die
        	b) Exp 2: 10,000 throws, 2 dice
        	c) Exp 3: 10,000 throws, 3 dice
        	d) Exp 4; 10,000 throws, 6 dice.

Show the distribution of results for each experiment. Which of these experiments produces a distribution most similar to a normal normal distribution? Justify your conclusion using Q-Q plots.

**Answer:** 

```r
# -------------------------------
# Question 4
# -------------------------------

#Roll with 1 die

exp1 = data.frame(length=round(runif(10000, min=.5, max=6.5)))
ggplot(exp1, aes(length)) + geom_histogram(bins=6) 
ggplot(exp1, aes(sample = length))+ stat_qq() + stat_qq_line() + ggtitle("Normal distribution with 1 die")



#Roll with 2 dice
exp2 = data.frame(length = round(runif(10000, min=.5, max=6.5)), length2 = round(runif(10000, min=.5, max=6.5))) 
exp2$sum = (exp2$length + exp2$length2)                  
                   
ggplot(exp2, aes(sum)) + geom_histogram(bins=10) 
ggplot(exp2, aes(sample=sum))+ stat_qq() + stat_qq_line() + ggtitle("Normal distribution with 2 dice")


#Roll with 3 dice  
exp3 <- data.frame(length=round(runif(10000, min=.5, max=6.5)),
                                length2=round(runif(10000, min=.5, max=6.5)),
                                        length3=round(runif(10000, min=.5, max=6.5)))
exp3$sum = (exp3$length + exp3$length2 + exp3$length3)
                     
ggplot(exp3, aes(sum)) + geom_histogram(bins=15) 
ggplot(exp3, aes(sample=sum))+ stat_qq() + stat_qq_line() + ggtitle("Normal distribution with 3 dice")
                   

#Roll with 6 dice
exp4 <- data.frame(length=round(runif(10000, min=.5, max=6.5)),
                                length2=round(runif(10000, min=.5, max=6.5)),
                                        length3=round(runif(10000, min=.5, max=6.5)),
                                                length4=round(runif(10000, min=.5, max=6.5)),
                                                        length5=round(runif(10000, min=.5, max=6.5)))
exp4$sum <- (exp4$length + exp4$length2 + exp4$length3 + exp4$length4 + exp4$length5)

ggplot(exp4, aes(sum)) + geom_histogram(bins=29) 
ggplot(exp4, aes(sample=sum))+ stat_qq() + stat_qq_line() + ggtitle("Normal distribution with 6 dice")


##The experiment 4 with 6 dice approximates the normal distribution better than the other experiments. 
#We can see in the histogram that experiment 4 approximates the normal curve the best due to the higher number of samples
# In the QQ plots, experiment 4 shows the closest approximation to the normal curve as shown by nearly diagonal line 
#comparing the values of the sample to what we would expect of a normal curve.
                                      
                                      



```
